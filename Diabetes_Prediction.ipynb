Importing the dependencies .

```python
import numpy as np
import pandas as pd
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from sklearn import svm
from sklearn.metrics import accuracy_score
```

Data Collection

PIMA Diabetes dataset

```python
# loading the dataset to a panda dataframe
diabetes_data = pd.read_csv('/content/diabetes.csv')
```

```python
diabetes_data.head()
```

```python
# number of rows & columns in dataset
diabetes_data.shape
```

```python
# getting the statistical information about dataset
diabetes_data.describe()
```

```python
diabetes_data['Outcome'].value_counts()
```

0 => Non-Diabetes

1 => Diabetes

```python
diabetes_data.groupby('Outcome').mean()
```

```python
# seperating the dataset and lables

A = diabetes_data.drop(columns = 'Outcome', axis = 1)
B = diabetes_data['Outcome']
```

```python
print(A)
```

```python
print(B)
```

Data Standadization [Pre-Processing]

```python
scaler = StandardScaler()
```

```python
scaler.fit_transform(A)
```

```python
standard_data = scaler.fit_transform(A)
```

```python
print (standard_data)
```

```python
A = standard_data
```

```python
print(A)
print(B)
```

```python
# Splitting data for training and test purpose
A_train, A_test, B_train, B_test = train_test_split(A, B, test_size = 0.25, stratify = B, random_state = 1)
```

```python
print(A.shape, A_train.shape, A_test.shape)
```

```python
# Training the Model
classifier = svm.SVC(kernel = 'linear')
```

```python
# Training the SVM classifier
classifier.fit(A_train, B_train)
```

Model Evaluation

Accuracy Score

```python
# Accuracy Score of Model on training data
A_train_prediction = classifier.predict(A_train)
training_data_accuracy = accuracy_score(A_train_prediction, B_train)
```

```python
print('Accuracy score of Model on training data : ', training_data_accuracy)
```

```python
# Accuracy Score of Model on Test data
A_test_prediction = classifier.predict(A_test)
test_data_accuracy = accuracy_score(A_test_prediction, B_test)
```

```python
print('Accuracy score of Model on test data : ', test_data_accuracy)
```

Making a Diabetes Prediction System

```python
input_data =(13,145,82,19,110,22.2,0.245,57)

# Changing the input_data to numpy array
input_data_as_numpy_array = np.asarray(input_data)

# Reshape the array as we are predicting for one instance
input_data_reshaped = input_data_as_numpy_array.reshape(1,-1)

# Standardize the input data
std_data = scaler.transform(input_data_reshaped)
print(std_data)

prediction = classifier.predict(std_data)
print(' \n prediction : ',prediction)

result = 'Not Diabetic' if prediction[0] == 0 else 'Diabetic'
print(f'\n Result : The Person is {result} \n')
```
